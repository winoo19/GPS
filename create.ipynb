```python
import pandas as pd
```

```python
cruces = pd.read_csv('data/cruces_clean.csv')
cruces.head()
```

```python
direcciones = pd.read_csv('data/direcciones_clean.csv')
direcciones.head()
```

```python
dists = set()
for i in range(len(cruces)):
    for j in range(len(cruces)):
        x1, y1 = cruces.iloc[i][['x', 'y']]
        x2, y2 = cruces.iloc[j][['x', 'y']]
        dists.add(((x1-x2)**2 + (y1-y2)**2)**0.5)
```

```python
def repl(dir):
    '''Replace each number in dir with its ascii code
    ej. hola 64 -> hola A'''
    import re
    return re.sub(r'\d+', lambda x: chr(int(x.group())), dir)


```
