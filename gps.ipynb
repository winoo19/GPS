```python
g._dirigido=False
G = g.convertir_a_NetworkX()
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G0 = G.subgraph(Gcc[0])
pos={k: k for k in g.adj.keys()}
k = Grafo()
k.NetworkX_a_grafo(G0)
aristas=k.kruskal()
G1 = k.kruskal_to_graph(aristas)
nx.draw(G0, pos=pos, node_size=0.1, arrows=False, width=0.1)
nx.draw(G1, pos=pos, edge_color="r", node_size=0.1, arrows=False, width=0.1)
```

```python
g = Grafo()
g.load_graph("plano_de_madrid.txt")
G = g.convertir_a_NetworkX()
'''random node'''
origen = list(G.nodes)[100]
destino = list(G.nodes)[-100]
'''shortest path'''
path = nx.shortest_path(G, origen, destino)
'''draw'''
nx.draw(G, pos={k: k for k in g.adj.keys()}, node_size=0.1, arrows=False, width=0.1)
nx.draw_networkx_nodes(G, pos={k: k for k in g.adj.keys()}, nodelist=path, node_color="r", node_size=0.1)
nx.draw_networkx_edges(G, pos={k: k for k in g.adj.keys()}, edgelist=[(path[i], path[i+1]) for i in range(len(path)-1)], edge_color="r", width=0.1)
```

```python
import networkx as nx
from grafo import Grafo
g = Grafo()
g.load_graph("plano_de_madrid.txt")
G = g.convertir_a_NetworkX()
'''random node'''
origen = list(G.nodes)[100]
destino = list(G.nodes)[-100]
'''shortest path'''
path = g.camino_minimo(origen, destino)
'''draw'''
nx.draw(G, pos={k: k for k in g.adj.keys()}, node_size=0.1, arrows=False, width=0.1)
nx.draw_networkx_nodes(G, pos={k: k for k in g.adj.keys()}, nodelist=path, node_color="r", node_size=0.1)
nx.draw_networkx_edges(G, pos={k: k for k in g.adj.keys()}, edgelist=[(path[i], path[i+1]) for i in range(len(path)-1)], edge_color="r", width=0.1)
```
