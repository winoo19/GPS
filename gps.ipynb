```python
from grafo import Grafo
import networkx as nx
```

```python
g = Grafo()
g.load_graph("plano_de_madrid.txt")
```

```python
G = g.convertir_a_NetworkX()
nx.draw(G, pos={k: k for k in g.adj.keys()}, node_size=0.1, arrows=False, width=0.1)
```

```python
g._dirigido=False
G = g.convertir_a_NetworkX()
Gcc = sorted(nx.connected_components(G), key=len, reverse=True)
G0 = G.subgraph(Gcc[0])
pos={k: k for k in g.adj.keys()}
k = Grafo()
k.NetworkX_a_grafo(G0)
aristas=k.kruskal()
G1 = k.kruskal_to_graph(aristas)
nx.draw(G0, pos=pos, node_size=0.1, arrows=False, width=0.1)
nx.draw(G1, pos=pos, edge_color="r", node_size=0.1, arrows=False, width=0.1)
```
