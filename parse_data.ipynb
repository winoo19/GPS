```python
import pandas as pd
```

## DIRECCIONES

```python
direcciones = pd.read_csv('data/direcciones.csv', sep=';', encoding='latin-1',
dtype = {'Coordenada X (Guia Urbana) cm': str, 'Coordenada Y (Guia Urbana) cm': str})
direcciones
```

### Direcciones Completas

```python
def parse_direccion_completa(direccion_completa):
    direccion_completa = direccion_completa.strip().replace(',', '')
    direccion_completa = ' '.join([w if not w.isnumeric() else str(int(w)) for w in direccion_completa.split()])
    return direccion_completa
direcciones["Direccion completa"] = direcciones["Direccion completa para el numero"].apply(parse_direccion_completa)
direcciones["Direccion completa"]
```

### Coordenadas en centimetros

```python
direcciones = direcciones[direcciones['Coordenada X (Guia Urbana) cm'].str.isnumeric()]
direcciones = direcciones[direcciones['Coordenada Y (Guia Urbana) cm'].str.isnumeric()]
direcciones['Coordenada X (Guia Urbana) cm'] = direcciones['Coordenada X (Guia Urbana) cm'].astype(int)
direcciones['Coordenada Y (Guia Urbana) cm'] = direcciones['Coordenada Y (Guia Urbana) cm'].astype(int)
direcciones.rename(columns={'Coordenada X (Guia Urbana) cm': 'x', 'Coordenada Y (Guia Urbana) cm': 'y'}, inplace=True)
direcciones.shape
```

```python
direcciones["Clase de la via"].value_counts()
```

### Quitar columnas sobrantes

```python
direcciones.columns
```

```python
direcciones.rename(columns={'Codigo de via':'id_via'}, inplace=True)
```

```python
direcciones=direcciones[['Direccion completa','id_via', 'x', 'y']]
direcciones
```

```python
direcciones.to_csv('data/direcciones_clean.csv', index=False)
```

## CRUCES

```python
cruces = pd.read_csv('data/cruces.csv', sep=';', encoding='latin-1')
cruces.shape
```

```python
cruces["Clase de la via tratado"].value_counts()
```

## Coordenadas en cm

```python
cruces.rename(columns={'Coordenada X (Guia Urbana) cm (cruce)': 'x', 'Coordenada Y (Guia Urbana) cm (cruce)': 'y'}, inplace=True)
```

## Columnas Irrelevantes

```python
cruces.rename(columns={"Codigo de vía tratado": "id_via"}, inplace=True)
cruces.rename(columns={"Codigo de via que cruza o enlaza": "id_via_cruzada"}, inplace=True)
cruces.rename(columns={"Literal completo del vial tratado": "nombre_via"}, inplace=True)
cruces.rename(columns={"Literal completo del vial que cruza": "nombre_via_cruzada"}, inplace=True)
cruces.rename(columns={"Clase de la via tratado": "tipo_via"}, inplace=True)
cruces["nombre_via"]=cruces["nombre_via"].apply(lambda x: x.strip().replace(',', ''))
cruces["nombre_via_cruzada"]=cruces["nombre_via_cruzada"].apply(lambda x: x.strip().replace(',', ''))
cruces=cruces[['id_via', 'nombre_via', 'id_via_cruzada', 'nombre_via_cruzada','tipo_via', 'x', 'y']].copy()
cruces[cruces["id_via"]==127]
```
