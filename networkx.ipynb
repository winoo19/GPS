```python
import networkx as nx
import matplotlib.pyplot as pltp
```

2.

```python
'''grafo no dirigido'''
G = nx.Graph()
V = [1,2,3,4,5]
A = [(1,2),(1,4),(3,4),(2,4),(3,2),(3,5)]
G.add_nodes_from(V)
G.add_edges_from(A)
```

```python
G.adj
```

3.

```python
'''delete vertex 5'''
G.remove_node(5)
```

4.

```python
nx.draw(G, with_labels=True)
```

5.

```python
'''grafo decorado'''
G2 = nx.Graph()
V = ["a","b",10,11,3.14] 
A = [("a",10),("a",3.14),("b",11)]

G2.add_nodes_from(V)
G2.add_edge("a",10, text="Una cadena", weight=1.1)
G2.add_edge("a",3.14, text="Otra cadena", weight=1.2)
G2.add_edge("b",11, text="Un objeto", weight=5.7)

'''draw weights'''
nx.draw(G2, with_labels=True)
```

6.

```python
'''grafo dirigido'''
Gd = nx.DiGraph()
V = [1,2,3,4]
A = [(1,2),(2,1),(1,4),(3,4),(2,4),(4,2),(3,2)]
Gd.add_nodes_from(V)
Gd.add_edges_from(A)
nx.draw(Gd, with_labels=True)
```

7.

```python
'''Grados de los nodos'''
print("Grados de los nodos")
print(G.degree())
print(G2.degree())
print(Gd.degree())
'''Grados entrantes y salientes'''
print("Grados entrantes y salientes")
print(Gd.in_degree())
print(Gd.out_degree())
```

8.

```python
'''componentes conexas'''
print("Componentes conexas")
print(list(nx.connected_components(G)))
print(list(nx.connected_components(G2)))
'''componentes fuertemente conexas'''
print("Componentes fuertemente conexas")
print(list(nx.weakly_connected_components(Gd)))
print(list(nx.strongly_connected_components(Gd)))
```

9.

```python
'''arbol abarcador minimo de G'''

T = nx.minimum_spanning_tree(G)

'''Draw G and T'''
pos = nx.spring_layout(G)
print(pos)
nx.draw(G, pos, with_labels=True)
nx.draw(T, pos, with_labels=True, edge_color='r')
'''node width smaller'''
```

10.

```python
'''draw shortest path in G'''
sp = nx.shortest_path(G, source=1, target=3)
sp_edges = list(zip(sp, sp[1:]))
nx.draw(G, pos, with_labels=True)
'''draw arrows'''
nx.draw_networkx_edges(G, pos, edgelist=sp_edges, edge_color='r')
```
